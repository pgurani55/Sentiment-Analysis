```python
#importing the datset
dataset=pd.read_csv('Restaurant_Reviews.tsv',delimiter='\t',quoting=3)
```

```python
dataset
```

```python
import string
import nltk
from nltk.corpus import stopwords
def textpreprocess(feature):
    #Remove Punctuations
    remove_punctuation = [char for char in feature if char not in string.punctuation] 
    #Character to sentences
    without_punctuation = "".join(remove_punctuation)
    #Sentences to words
    words = without_punctuation.split(' ')
    #Character without Stopwords
    without_stopwords = [word for word in words if word.lower() not in stopwords.words('english')]
    return without_stopwords
```

```python
output = textpreprocess(dataset['Review'])
output
```

```python
#Now the Preprocessing has already been done!
#Creating Bag of Words!
from sklearn.feature_extraction.text import CountVectorizer
word_vector = CountVectorizer(analyzer=textpreprocess)
final_words = word_vector.fit(dataset['Review'])
```

```python
final_words.vocabulary_
```

```python

bag_of_words=word_vector.transform(dataset['Review'])
```

```python
bag_of_words
```

```python
#Now just Bag of Words is not sufficient so now apply TF-IDF ( Term Freq / Document Freq )
from sklearn.feature_extraction.text import TfidfTransformer
tf_idf = TfidfTransformer().fit(bag_of_words)
```

```python
final_tf = tf_idf.transform(bag_of_words)
final_tf#This is the output of TF_IDF
```

```python
pd.DataFrame(final_tf.toarray()[0]) #So this is the Division of TermFreq/DocumFreq
```

```python
#Now we can apply to the ML Model
from sklearn.neighbors import KNeighborsClassifier
baseline_model = KNeighborsClassifier()
baseline_model.fit(final_tf,dataset['Liked'])
```

```python
baseline_model.score(final_tf,dataset['Liked'])
```

```python
from sklearn.metrics import confusion_matrix,classification_report
clas = confusion_matrix(dataset['Liked'],baseline_model.predict(final_tf))
```
